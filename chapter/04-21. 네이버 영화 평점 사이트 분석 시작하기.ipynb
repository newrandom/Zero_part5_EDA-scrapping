# 영화 평점순 랭킹 주소
https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20220801

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from urllib.request import urlopen

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20180315'
page = urlopen(url)

soup = BeautifulSoup(page, 'html.parser')
soup
```

# 제목과 평점 가져오기


```python
soup.find_all('div',class_='tit5')      # 제목
```

```python
soup.find_all('div', {'class':'tit5'})[0].a
```

```python
soup.find_all('div', 'tit5')[0].a.get_text()        # 최종 영화 제목 가져오기 끝
```

영화 평점 가져오기

```python
soup.find_all('td','point')
```

```python
soup.find_all('td','point')[0].string       # 평점도 가져옴.
```

### 종합

```python
# 영화 이름 전체 가져오기
end = len(soup.find_all('div','tit5'))
movie_name = [soup.find_all('div','tit5')[n].a.string for n in range(0, end)]
movie_name
```

```python
# 영화 평점 전체 가져오기
end = len(soup.find_all('td', 'point'))
movie_point = [soup.find_all('td','point')[n].string for n in range(0, end)]
movie_point
```

```python
# 일자별 영화이름, 평점 받아오기
from tqdm import tqdm
from urllib.request import Request, urlopen
import time

movie_date = []
movie_name = []
movie_point= []

for today in tqdm(date):
    html = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={date_}'
    # response = urlopen(html.format(date_=today.strftime('%Y%m%d')), )
    req = Request(html.format(date_=today.strftime('%Y%m%d')), headers={'User-Agent' : 'Chrome'})
    response = urlopen(req)
    soup = BeautifulSoup(response, 'html.parser')

    end = len(soup.find_all('td', 'point'))

    movie_date.extend([today for _ in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])

    time.sleep(0.3)
```
