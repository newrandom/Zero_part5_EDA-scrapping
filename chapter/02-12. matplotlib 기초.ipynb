```python
import matplotlib.pyplot as plt
```

```python
def drawGraph() :
    plt.figure(figsize= (10, 6))
    plt.plot(t, np.sin(t), label='sin')
    plt.plot(t, np.cos(t), label='cos')
    plt.grid()
    plt.legend()
    plt.xlabel('time')
    plt.ylabel('Amplitude')
    plt.title('Example of sinewave')
    plt.show()
```

```python
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]

def drawGraph():
    plt.figure(figsize=(10,6))
    plt.plot(
        t, y, color = 'green',
        linestyle='dashed',
        marker='o', markerfacecolor='blue',
        markersize=12,
    )

    plt.xlim([-0.5, 6.5])
    plt.ylim([0.5, 9.5])
    plt.show()

drawGraph()
```

### 예제1 : 그래프 기초

#### 삼각함수 그리기
- np.arange(a, b, s) : a부터 b 까지 s 라는 간격으로 데이터를 만들어줌
- np.sin(value)


```python
import numpy as np

t = np.arange(0, 12, 0.01)
y = np.sin(t)
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t))
plt.plot(t, np.cos(t))
plt.show()
```

- 1 . 격자무늬 추가
- 2 . 그래프 제목 추가
- 3 . x축, y축 제목 추가
- 4 . 주황색, 파란색 선 데이터의 의미 구분

```python
def drawGraph() :
        
    plt.figure(figsize=(10,6))
    plt.plot(t, np.sin(t), label = 'sin')
    plt.plot(t, np.cos(t), label = 'cos')
    plt.grid(True)
    plt.legend(loc='upper right')    #labels=['sin','cos'])        # legend 안에 labels를 차례로 넣는 방법도 있다.
    plt.title('Example of sinewave')
    plt.xlabel('time')
    plt.ylabel('Amplitude(진폭)')
    plt.show()
```

```python
drawGraph()
```
