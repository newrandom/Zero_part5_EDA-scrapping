```python
import pandas as pd
import numpy as np
import set_matplotlib_hangul

%matplotlib inline
```

```python
population = pd.read_excel('../data_08/07_population_raw_data.xlsx', header=1)
population.fillna(method='pad', inplace =True)
population.head()
```

```python
population.columns
```

```python
population.rename(
    columns={'행정구역(동읍면)별(1)' : '광역시도' , '행정구역(동읍면)별(2)' : '시도', '계':'인구수'}, inplace=True
)
population.head()
```

```python
# 시도 컬럼의 소계 데이터 삭제
population=population[(population['시도'] != '소계')]
population.head()
```

```python
population['항목'].unique()
```

```python
# 항목의 총인구수(명) , 남자인구수(명), 여자인구수(명) 을 줄여서 표현하기
population.is_copy = False

population.rename(columns={'항목': '구분'}, inplace=True)

population.loc[population['구분'] == '총인구수 (명)', '구분']  = '합계'
population.loc[population['구분'] == '남자인구수 (명)', '구분']  = '남자'
population.loc[population['구분'] == '여자인구수 (명)', '구분']  = '여자'

population.head()
```

```python
population.columns
```

```python
# 나이 컬럼의 범위 축소
population['20-39세'] = (
    population['20 - 24세']
    + population['25 - 29세']
    + population['30 - 34세']
    + population['35 - 39세']
)

population['65세 이상'] = (
    population['65 - 69세']
    + population['70 - 74세']
    + population['75 - 79세']
    + population['80 - 84세']
    + population['85 - 89세']
    + population['90 - 94세']
    + population['95 - 99세']
    + population['100+']
)

population.head(5)
```

```python
# pivot_table 이용하여 인덱스, 컬럼 변경하기
pop = pd.pivot_table(
    population, index=['광역시도','시도'], columns=['구분'], values=['인구수', '20-39세', '65세 이상']
)
pop
```

```python
# 소멸인구 비율 보기
## 인구소멸 비율이 절반임 (?)
pop['소멸비율'] = pop['20-39세', '여자'] / (pop['65세 이상', '합계'] / 2)
pop.head()
```

```python
pop['소멸위기지역'] = pop['소멸비율'] < 1.0
pop.head()
```

```python
# 소멸 위기지역 조회
pop[pop['소멸위기지역'] == True].index.get_level_values(1)
```

```python
# index 초기화 하기
pop.reset_index(inplace=True)
pop.head()
```

```python
tmp_columns = [
    pop.columns.get_level_values(0)[n] + pop.columns.get_level_values(1)[n]
    for n in range(0, len(pop.columns.get_level_values(0)))
]
pop.columns = tmp_columns
pop.head()
```

```python
pop.to_csv('../data_08/population.csv', sep=',', encoding='utf-8')
```
