### 02. excel 불러오기

```python
# xls 파일 불러오기
pop_seoul = pd.read_excel('../data/01. Seoul_Population.xls', header=2, usecols=['자치구','계', '계.1','계.2','65세이상고령자'])
pop_seoul.head()
```

```python
pop_seoul.rename(columns={
    pop_seoul.columns[0] : '구별',
    pop_seoul.columns[1] : '인구수',
    pop_seoul.columns[2] : '한국인',
    pop_seoul.columns[3] : '외국인',
    pop_seoul.columns[4] : '고령자'}, inplace=True)
```

```python
pop_seoul.head()
```

```python
# 오름차순으로 정렬
cctv_seoul.sort_values(by='소계', ascending=True).head(5)
```

```python
# 내림차순으로 정렬
cctv_seoul.sort_values(by='소계', ascending=False).head(5)
```

```python
# 기존 컬럼이 없으면 추가, 있으면 수정
cctv_seoul['최근증가율'] = (
    (cctv_seoul['2016년'] + cctv_seoul['2015년'] + cctv_seoul['2014년']) / cctv_seoul['2013년도 이전'] * 100
)

cctv_seoul.head()
```

```python
cctv_seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

# 인구현황 데이터 훑어보기

```python
pop_seoul.head()
```

```python
pop_seoul.tail()
```

```python
pop_seoul.drop([0], axis=0, inplace=True)       # 합계 row 없앰
pop_seoul.head()
```

```python
len(pop_seoul['구별'].unique())
```

```python
## 외국인 비율, 고령자 비율 컬럼 만들기
pop_seoul['외국인비율'] = pop_seoul['외국인'] / pop_seoul['인구수'] * 100
pop_seoul['고령자비율'] = pop_seoul['고령자'] / pop_seoul['인구수'] * 100
pop_seoul.head()
```

```python
pop_seoul.sort_values(by='인구수', ascending=False).head()
                        # by 대신에 ['인구수'] 를 써도 된다.
```

```python
pop_seoul.sort_values(['외국인'], ascending=False).head(5)
```

```python
pop_seoul.sort_values(by='외국인비율', ascending=False).head()
```

```python
pop_seoul.sort_values(by='고령자', ascending=False).head()
```

```python
pop_seoul.sort_values(by='고령자비율', ascending=False).head()
```
