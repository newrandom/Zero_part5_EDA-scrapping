## pandas 로 변환


```python
import pandas as pd

def get_fields(json_data):
    title = [each['title'] for each in json_data['items']]
    link = [each['link'] for each in json_data['items']]
    lprice = [each['lprice'] for each in json_data['items']]
    mall_name = [each['mallName'] for each in json_data['items']]

    result_pd = pd.DataFrame({
        'title' : title,
        'lprice' : lprice,
        'link': link,
        'mall' : mall_name
    }, columns=['title', 'lprice', 'link', 'mall'])
    return result_pd
```

### <b> 태그 없애기

```python
def delete_tag(input_str):
    input_str = input_str.replace('<b>', '')
    input_str = input_str.replace('</b>', '')
    return input_str
```

```python
import pandas as pd

def get_fields(json_data):
    title = [delete_tag(each['title']) for each in json_data['items']]      # 태그 없애기 함수를 바로 반영
    link = [each['link'] for each in json_data['items']]
    lprice = [each['lprice'] for each in json_data['items']]
    mall_name = [each['mallName'] for each in json_data['items']]

    result_pd = pd.DataFrame({
        'title' : title,
        'lprice' : lprice,
        'link': link,
        'mall' : mall_name
    }, columns=['title', 'lprice', 'link', 'mall'])
    return result_pd
```

```python
get_fields(result)
```

## 모든 데이터 가져오기

```python
import json

result_mol = []
for n in range(1, 1000, 100):
    result = json.loads(naver_api(id = id, pw = pw, search = search, where = where, start_num = n, disp_num=100))       # 1 ~ 100 데이터 > json.loads
    pd_result = get_fields(result)      # json_data 를 데이터 정리하여 PD로 뽑기

    result_mol.append(pd_result)        # df 데이터를 리스트에 넣기

result_mol = pd.concat(result_mol)      # concat : 공통된 컬럼에 모두 집어넣음.
```

```python
result_mol.info()       # 인덱스 번호가 0~99로 끊겨서 100번 반복됨. >> 고칠것.
```

```python
result_mol.reset_index(drop=True, inplace=True)
result_mol.info
```

```python
result_mol.head()
```

```python
# 가격은 숫자형(float)으로
result_mol['lprice'] = result_mol['lprice'].astype('float')
result_mol.info()
```

---
## Excel로 저장하기

```python
writer = pd.ExcelWriter('../data_07/molskin_diary_in_naver_shop.xlsx', engine='xlsxwriter')
result_mol.to_excel(writer, sheet_name = 'Sheet1')

workbook = writer.book
worksheet =writer.sheets['Sheet1']
worksheet.set_column('A:A', 4)
worksheet.set_column('B:B', 60)
worksheet.set_column('C:C', 10)
worksheet.set_column('D:D', 10)
worksheet.set_column('E:E', 50)
worksheet.set_column('F:F', 10)

worksheet.conditional_format('C2:C1001', {'type' : '3_color_scale'})
writer.save()
```
