```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2 * np.sin(x + 0.5)
y3 = 3 * np.sin(x + 1.0)
y4 = 4 * np.sin(x + 1.5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3 ,x, y4)
# plt.plot(x, y1)
# plt.plot(x, y2)
# plt.plot(x, y3)
# plt.plot(x, y4)
plt.show()
```

```python
sns.set_style('white')
                # {darkgrid, whitegrid, dark, white, ticks}
plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
sns.despine()
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=tips['total_bill'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set3')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x='day', y='total_bill', data=tips, color='.5')
plt.show()
```

```python
# boxplot과 swarmplot 콜라보
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='.1')
plt.show()
```

```python
# total bill 과 tip 사이의 관계 파악
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, height=7)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, height=7)
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights.head(5)
```

```python
flights = flights.pivot('month', 'year', 'passengers')
flights.head(5)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt='d', cmap='YlGnBu')
plt.show()
```

```python
sns.lmplot(
    x='x',
    y='y',
    data=anscombe.query("dataset=='II'"),
    ci=None,
    scatter_kws={'s':80},
    size=7,
    order=2
)
plt.show()
```
