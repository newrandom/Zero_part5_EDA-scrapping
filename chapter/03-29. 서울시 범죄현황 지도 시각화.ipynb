```python
import pandas as pd
import json
```

```python
crime_anal_norm = pd.read_csv('../data2/02. crime_in_Seoul_final.csv', index_col='구별', encoding='utf-8')
geo_path = '../data2/02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
import folium
```

```python
import numpy as np
```

```python
# 경찰서별 정보를 가지고 범죄발생과 함께 정리

crime_anal_station=pd.read_csv(
    '../data2/02. crime_in_Seoul_1st.csv', index_col=0, encoding='utf-8'
)
col = ['살인검거', '강도검거', '강간검거', '절도검거', '폭력검거']
tmp = crime_anal_station[col] / crime_anal_station[col].max()
crime_anal_station['검거'] = np.mean(tmp, axis=1)       # 평균 검거 컬럼
crime_anal_station.head()
```

```python
# 경찰서 위치를 지도에 표시
m = folium.Map(location=[37.5502, 126.982], zoom_start=11)

for idx, row in crime_anal_station.iterrows():
    folium.Marker([row['lat'], row['lng']]).add_to(m)

m
```

```python
# 검거 컬럼의 데이터를 통한 경찰서별 범위
m = folium.Map(location=[37.5502, 126.982], zoom_start=11)

for idx, row in crime_anal_station.iterrows():
    folium.CircleMarker(
        [row['lat'], row['lng']],
        radius = row['검거'] * 50,
        popup = row['구분'] + " : " + '%.2f' % row['검거'],
        color = '#3186cc',
        fill=True,
        fill_color='#3186cc',
    ).add_to(m)

m
```

```python
# 구별 범죄 현황과 경찰서별 검거율을 함께 표시
m = folium.Map(location=[37.5502, 126.982], zoom_start=11)
m.choropleth(
    geo_data = geo_str,
    data=crime_anal_norm['범죄'],
    columns = [crime_anal_norm.index, crime_anal_norm['범죄']],
    fill_color='PuRd',
    key_on='feature.id',
    fill_opacity=.7,
    line_opacity=.2,
)

for idx, rows in crime_anal_station.iterrows():
    folium.CircleMarker(
        [rows['lat'], rows['lng']],
        radius=rows['검거'] * 50,
        popup = rows['구분'] + ' : ' + '%.2f' % rows['검거'],
        color='#3186cc',
        fill=True,
        fill_color='#3186cc',
    ).add_to(m)

m
```
