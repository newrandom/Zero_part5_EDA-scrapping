```python
import pandas as pd
import numpy as np
```

```python
crime_raw_data = pd.read_csv('../data2/02. crime_in_Seoul.csv', thousands=',', encoding='euc-kr')
crime_raw_data
```

```python
crime_raw_data = crime_raw_data[crime_raw_data['죄종'].notnull()]
crime_raw_data.head()
```

```python
crime_raw_data.tail()
```

---

# Pivot_table 사용하기

```python
crime_station = crime_raw_data.pivot_table(
    crime_raw_data, 
    index='구분', 
    columns=['죄종','발생검거'], 
    aggfunc=[np.sum]
)
crime_station.head()
```

```python
crime_station.columns
```

```python
crime_station['sum', '건수','강도','검거'][:5]
```

```python
crime_station.columns=crime_station.columns.droplevel([0,1])        # 다중 컬럼에서 특정 컬럼 제거
crime_station.head()
```
