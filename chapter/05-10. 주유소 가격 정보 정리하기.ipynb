```python
import pandas as pd
from glob import glob
```

```python
len(glob('../oil_station/지역_*.xls')), glob('../oil_station/지역_*.xls')
```

```python
stations_files = glob('../oil_station/지역_*.xls')
stations_files
```

```python
tmp = pd.read_excel(stations_files[0], header=2)
tmp.head()
```

```python
tmp_raw=[]
for file_name in stations_files:
    tmp = pd.read_excel(file_name, header= 2)
    tmp_raw.append(tmp)     # DataFrame이 list에 저장됨.

station_raw = pd.concat(tmp_raw)        # 구조가 동일한 DataFrame은 list 형태로 묶어서 concat 하면 된다.
```

```python
station_raw.info()
```

```python
station_raw.head()
```

```python
stations = pd.DataFrame(
    {
        '상호' : station_raw['상호'],
        '주소' : station_raw['주소'],
        '가격' : station_raw['휘발유'],
        '셀프' : station_raw['셀프여부'],
        '상표' : station_raw['상표'],
    }
)
stations.head()
```

```python
stations['구'] = [eachAddress.split()[1] for eachAddress in stations['주소']]
stations.head()
```

```python
stations[
'구'
].unique()      # 강서구 , 강서 
```

```python
len(stations['구'].unique())
```

```python
stations[stations['구'] == '강서']
```

```python
stations.loc[stations['구'] == '강서', '구'] = '강서구'
stations['구'].unique()
```

```python
stations.info()
```

```python
# stations['가격'] = stations['가격'].astype('float')
```

```python
stations[stations['가격'] == '-']       # 가격정보가 없는 주유소가 있음
```

```python
stations = stations[stations['가격'] != '-']        # 가격 정보가 있는 곳만 가져옴
stations.head()
```

```python
stations['가격'] = stations['가격'].astype('float')
stations.info()
```
